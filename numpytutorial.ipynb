create a numpy array of ones and a numpy array of zeroes

```python
import numpy as np
```

```python
c=np.ones((2,2))
d=np.zeros((2,2))
print("\n.Array of ones(2*2): \n", c)
print("\n.Array of zeroes: \n" , d)
```

create a random numpy array of size(6*3) and transpose it

```python
e = np.random.random((6,3))
transposed = e.T
print("\n6*3 Random Array: \n", e)
print("\nTransposed (3*6): \n", transposed)
```

Create two random numpy array of size (n*p) and (p*m) then compute the dot products

```python
n, p, m = 2,3,4
F = np.random.random((n,p))
G = np.random.random((p,m))
dotproducts = np.dot(F, G)

print("\nF (n * p)\n", F)
print("\nG (p * n)\n" , G)
print("\nDot Products:\n" , dotproducts)
```
